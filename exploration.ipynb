```r
library(ggplot2)
```

```r
# load the dataset
data <- read.csv("final.csv")
head(data)
```

```r
# extract each country's data for comparison
australia <- data[data$country == "Australia",]
uk <- data[data$country == "United Kingdom",]
argentina <- data[data$country == "Argentina",]
brazil <- data[data$country == "Brazil",]
denmark <- data[data$country == "Denmark",]
japan <- data[data$country == "Japan",]
korea <- data[data$country == "Korea",]
malaysia <- data[data$country == "Malaysia",]
mexico <- data[data$country == "Mexico",]
newzealand <- data[data$country == "New Zealand",]
singapore <- data[data$country == "Singapore",]
sweden <- data[data$country == "Sweden",]
switzerland <- data[data$country == "Switzerland",]
```

```r
# plot Australia's gdp per capita over time
ggplot(australia, aes(x=year, y=gdp_per_capita)) + 
geom_line() + 
ggtitle("Australia's GDP per Capita Over Time") + 
xlab("Year") + 
ylab("GDP per Capita") + 
theme_minimal()
```

```r
# plot each country's gdp per capita over time on the same plot
ggplot(data, aes(x=year, y=gdp_per_capita, color=country)) +
geom_line() +
ggtitle("GDP per Capita Over Time by Country") +
xlab("Year") +
ylab("GDP per Capita") +
# make australia's line darker for emphasis
scale_color_manual(values=c("Australia"="red", "United Kingdom"="blue", "Argentina"="green", 
"Brazil"="purple", "Denmark"="orange", "Japan"="black", "Korea"="brown", "Malaysia"="pink", 
"Mexico"="gray", "New Zealand"="cyan", "Singapore"="magenta", "Sweden"="yellow", "Switzerland"="darkgreen")) +
theme_minimal()
```

```r
# plot the national income of each country
ggplot(data, aes(x=year, y=adj_national_income, color=country)) +
geom_line(data=data[data$country != "Japan" & data$country != "United Kingdom",]) +
ggtitle("Adjusted National Income Over Time by Country") +
xlab("Year") +
ylab("Adjusted National Income") +
# make australia's line darker for emphasis
scale_color_manual(values=c("Australia"="red", "United Kingdom"="blue", "Argentina"="green", 
"Brazil"="purple", "Denmark"="orange", "Japan"="black", "Korea"="brown", "Malaysia"="pink", 
"Mexico"="gray", "New Zealand"="cyan", "Singapore"="magenta", "Sweden"="yellow", "Switzerland"="darkgreen")) +
theme_minimal()
```

```r
# plot the exports percentage of each country
ggplot(data, aes(x=year, y=trade_percentage, color=country)) +
# remove Singapore and Malaysia
geom_line(data=data[data$country != "Singapore" & data$country != "Malaysia",]) +
ggtitle("Trade Percentage Over Time by Country") +
xlab("Year") +
ylab("Trade Percentage") +
# make australia's line darker for emphasis
scale_color_manual(values=c("Australia"="red", "United Kingdom"="blue", "Argentina"="green", 
"Brazil"="purple", "Denmark"="orange", "Japan"="black", "Korea"="brown", "Malaysia"="pink", 
"Mexico"="gray", "New Zealand"="cyan", "Singapore"="magenta", "Sweden"="yellow", "Switzerland"="darkgreen")) +
theme_minimal()
```
